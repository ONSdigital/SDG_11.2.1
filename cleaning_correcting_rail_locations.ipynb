```python
corrections_to_codes = pd.read_csv("corrections_stations_not_in_locs.csv")
```

```python
corrections_to_codes.notes.unique()
```

```python
corrections_to_locs = pd.read_csv("corrections_locations_not_in_codes.csv")
corrections_to_locs.action.unique()
```

```python
corrections_to_locs.action.unique()
```

```python
# Delete erroneous stations from locations
stations_to_be_deleted = corrections_to_locs.action == 'delete from locs'
corrections_to_locs.rename(columns={" ": "station_code"}, inplace=True)
station_codes_to_be_deleted = corrections_to_locs.station_code[stations_to_be_deleted]
```

```python
# Get index of records to be deleted in station locations
locs_to_delete = station_locs_df.where(station_locs_df.station_code.isin(station_codes_to_be_deleted))
locs_to_delete.dropna(inplace=True)
index_to_delete = locs_to_delete.index
```

```python
# delete those records
station_locs_df.drop(index_to_delete, inplace=True)
```

```python
# Replace incorrect codes
stations_with_corrected_codes = corrections_to_locs.action == 'changed code'
stn_names_corrected_codes = corrections_to_locs[stations_with_corrected_codes]
stn_names_corrected_codes
```

```python
#Tidy up replace the old records
# stn_names_corrected_codes.drop(columns=["notes"], inplace=True)
just_names = stn_names_corrected_codes.station_name
```

```python
# Match on names (because codes won't match now they've been corrected)
name_match = station_locs_df.station_name.isin(just_names)
# Find the rows to delete with a namematch
records_to_delete_ = station_locs_df.where(name_match)
records_to_delete_.dropna(inplace=True)
```

```python
# Get the index numbers of those records, is this is what .drop takes
index_to_delete_2 = records_to_delete_.index
station_locs_df.drop(index_to_delete_2,inplace=True)
```

```python
# Drop the action col as this was only needed for developer use
station_locs_df.drop(columns=["action"], inplace=True)
```

```python
# Reset index 
stn_names_corrected_codes.reset_index(inplace=True)
```

```python
# As with station codes, correct the col name on the locations df here
stn_names_corrected_codes.rename(columns={" ": "station_code"}, inplace=True)
```

```python
# Drop the useless index and action cols
stn_names_corrected_codes.drop(columns=["index", "action"], inplace=True)
stn_names_corrected_codes.head()
```

```python
# Shape the original locs df in the same way 
station_locs_df = station_locs_df[["station_code",	"station_name",	"latitude",	"longitude"]]
```

```python
# Check how it looks
station_locs_df.head()
```

```python
# Now the ones to be replaced have been deleted, add them back in
station_locs_df = pd.concat([station_locs_df, stn_names_corrected_codes])
station_locs_df
```

```python
# Checking for duplicates
station_locs_df.index.duplicated()
```

```python
station_locs_df.set_index("station_code", inplace=True)
```

```python
# Final push
# Drop records where the index (station code) is duplicated 
station_locs_df = station_locs_df[~station_locs_df.index.duplicated(keep="first")]
# Drop records where index is na (some of the dupes have an na index)
station_locs_df = station_locs_df[~station_locs_df.index.isna()]
# Change the datatypes to float in the lat and lon
station_locs_df.latitude = station_locs_df.latitude.astype('float')
station_locs_df.longitude = station_locs_df.longitude.astype('float')
# Check - looking good
station_locs_df.head(20)
```

```python
# Output results as required 
station_locs_df.to_json("all_uk_stations_locations.json", orient="index")
station_locs_df.to_csv("all_uk_stations_locations.csv")
```
